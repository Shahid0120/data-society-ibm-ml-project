# Creating Time Diff values between planned/actualy start/end dates

```python
from datetime import datetime

# Create a function which converts to date integer find differen and places this into an array 
def timeDiff(df, col1, col2):
    date_time_diff = []
    for index, value in enumerate(df[col1]):
        # checks if is in the format %m/%d/%Y else just makes difference 0 
        try:
            start_index_date = datetime.strptime(df[col1][index], '%m/%d/%Y')
            end_index_date = datetime.strptime(df[col2][index], '%m/%d/%Y')

            # Find the difference
            date_diff = end_index_date - start_index_date
            date_diff = date_diff.days
        except:
            # should i do this as X?
            date_diff = Nan
        date_time_diff.append(date_diff)
        
    return date_time_diff


# df['time_diff_planned_start_planned_end']
time_diff_actual_start_planned_end_arr = timeDiff(df, 'project_phase_actual_start_date', 'project_phase_planned_end_date') 
time_diff_actual_start_actual_end_arr = timeDiff(df, 'project_phase_actual_start_date', 'project_phase_actual_end_date') 
time_diff_actual_end_date_minus_planned_end_date_arr = timeDiff(df, 'project_phase_planned_end_date' , 'project_phase_actual_end_date',)

# add to existing dataframe arr 

df['time_diff_actual_end_date_minus_actual_start_date'] = time_diff_actual_start_actual_end_arr
df['time_diff_planned_end_date_minus_actual_start_date'] = time_diff_actual_start_planned_end_arr

# important negative values actual_end < planned (so it took less time then expective) while positve actual_end > planned (took more time then expected)
df['time_diff_actual_end_date_minus_planned_end_date'] = time_diff_actual_end_date_minus_planned_end_date_arr

df.head()
```
