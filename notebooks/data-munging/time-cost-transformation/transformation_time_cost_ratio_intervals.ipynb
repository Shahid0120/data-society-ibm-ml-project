# Creating Time Diff values between planned/actualy start/end dates

```python
import datetime
from datetime import datetime

# Create a function which converts to date integer find differen and places this into an array 
def timeDiff(df, col1, col2):
    date_time_diff = []
    for index, value in enumerate(df[col1]):
        # checks if is in the format %m/%d/%Y else just makes difference 0 
        try:
            start_index_date = datetime.strptime(df[col1][index], '%m/%d/%Y')
            end_index_date = datetime.strptime(df[col2][index], '%m/%d/%Y')

            # Find the difference
            date_diff = end_index_date - start_index_date
            date_diff = date_diff.days
        except:
            # should i do this as X?
            date_diff = 'Nan'
        date_time_diff.append(date_diff)
        
    return date_time_diff


# df['time_diff_planned_start_planned_end']
time_diff_actual_start_planned_end_arr = timeDiff(df, 'project_phase_actual_start_date', 'project_phase_planned_end_date') 
time_diff_actual_start_actual_end_arr = timeDiff(df, 'project_phase_actual_start_date', 'project_phase_actual_end_date') 
time_diff_actual_end_date_minus_planned_end_date_arr = timeDiff(df, 'project_phase_planned_end_date' , 'project_phase_actual_end_date',)

# add to existing dataframe arr 

df['time_diff_actual_end_date_minus_actual_start_date'] = time_diff_actual_start_actual_end_arr
df['time_diff_planned_end_date_minus_actual_start_date'] = time_diff_actual_start_planned_end_arr

# important negative values actual_end < planned (so it took less time then expective) while positve actual_end > planned (took more time then expected)
df['time_diff_actual_end_date_minus_planned_end_date'] = time_diff_actual_end_date_minus_planned_end_date_arr

df.head()
```

## Adding Time Ratio 
Schedule Performance Index : (Actual_end - Actual_start ) / (planned - actual_start) : 1> exactly as planned <1 behind planned
Schedule Variance : (Actual_end - Planned_end_date) / (planned_end_data - actual_start_date) : 0> ahead of scedule <0 behind scedule

```python
import datetime
from datetime import datetime
import pandas as pd

def calculateTimeRatio(df, actual_start_date_col, planned_end_date_col, actual_end_date_col):
    schedule_performance_arr = []
    schedule_variance_arr = []
    
    for index in range(len(df)):
        try:
            # Parsing dates
            actual_start_date = datetime.strptime(df[actual_start_date_col].iloc[index], '%m/%d/%Y')
            planned_end_date = datetime.strptime(df[planned_end_date_col].iloc[index], '%m/%d/%Y')
            actual_end_date = datetime.strptime(df[actual_end_date_col].iloc[index], '%m/%d/%Y')

            # Schedule Ratio calculations
            schedule_performance = (actual_end_date - actual_start_date) / (planned_end_date - actual_start_date)
            schedule_variance = (actual_end_date - planned_end_date) / (planned_end_date - actual_start_date)

            # Append to arrays
            schedule_performance_arr.append(schedule_performance)
            schedule_variance_arr.append(schedule_variance)

        except:
            schedule_performance_arr.append('Nan')
            schedule_variance_arr.append('Nan')
            continue
    
    return schedule_performance_arr, schedule_variance_arr

# Example usage
schedule_performance_arr, schedule_variance_arr = calculateTimeRatio(
    df,
    'project_phase_actual_start_date',
    'project_phase_planned_end_date',
    'project_phase_actual_end_date'
)

df['schedule_performance_ratio'] = schedule_performance_arr
df['schedule_variance_ratio'] = schedule_variance_arr
```

# Cost Ratio 


```python
import pandas as pd
import numpy as np
from numpy import divide

def calculateFinancialRatios(df, final_estimate_col, project_budget_col, actual_spending_col):
    cpi_arr = []
    ber_arr = []
    cv_arr = []
    
    for index in range(len(df)):
        try:
            # Get the values from the DataFrame
            final_estimate = df[final_estimate_col].iloc[index]
            actual_spending = df[actual_spending_col].iloc[index]

            # stop inf 
            if int(actual_spending) == 0:
                cpi_arr.append('Not recorded yet')
            else:  
                # Calculate Cost Performance Index (CPI)
                cpi = int(final_estimate) / int(actual_spending)
                cpi_arr.append(cpi)
        except:
            cpi_arr.append('Nan')
            
        try:
            # Try to convert values
            project_budget = df[project_budget_col].iloc[index]
            actual_spending = df[actual_spending_col].iloc[index]
        
            # stop inf 
            if int(actual_spending) == 0:
                ber_arr.append('Not recorded yet')
            else:
                # Budget Efficiency Ratio

                ber = int(project_budget) / int(actual_spending)
                ber_arr.append(ber)
        except:
            ber_arr.append('Nan')
        try:
            # Try to convert values
            project_budget = df[project_budget_col].iloc[index]
            final_estimate = df[final_estimate_col].iloc[index]
            actual_spending = df[actual_spending_col].iloc[index]


            # stop inf 
            if int(project_budget) == 0:
                cv_arr.append('Not recorded yet')
            else:
                 # Calculate Cost Variance (CV)
                cv = (int(final_estimate) - int(actual_spending)) / int(project_budget)
                
                cv_arr.append(cv)
        except:
            cv_arr.append('Nan')
    return cpi_arr, ber_arr, cv_arr


cpi_arr, ber_arr, cv_arr = calculateFinancialRatios(
    df,
    'final_estimate_of_actual_costs_through_end_of_phase_amount',
    'project_budget_amount',
    'total_phase_actual_spending_amount'
)


df['cost_performance_index'] = cpi_arr
df['budget_efficiency_ratio'] = ber_arr
df['cost_variance'] = cv_arr
```
